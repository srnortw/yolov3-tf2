```python
import fiftyone as fo
fo.delete_datasets("afric*")
# fo.delete_datasets("sea*")
# fo.delete_datasets("indoor*")
fo.delete_datasets("const*")
os.chdir('/content/drive/MyDrive/yolov3-tf2')
```

```python
import os

dataset_root = "testx"
splits = ["traincv", "cv", "test"]
#splits = ["traincv"]
#splits = ["cv", "test"]

labelx=["buffalo", "elephant", "rhino","zebra"]
#labelx=["fish", "jellyfish", "penguin","puffin", "shark", "starfish","stingray"]
# labelx=['door', 'cabinetDoor', 'refrigeratorDoor', 'window', 'chair', 'table', 'cabinet', 'couch', 'openedDoor', 'pole']
#labelx=['helmet','no-helmet','no-vest','person','vest']


dataset = fo.Dataset("african-animals-datasetx")
# dataset = fo.Dataset("indoors-objdet-dataset")
#dataset = fo.Dataset("construction-safety-objdet-dataset")

for split in splits:
    img_dir = os.path.join(dataset_root, "images", split)
    pred_dir = os.path.join(dataset_root, "predictions", split)
    lbl_dir = os.path.join(dataset_root, "labels", split)

    # # Load just the image paths
    # samples = fo.Dataset.from_dir(
    #     dataset_dir=img_dir,
    #     dataset_type=fo.types.ImageDirectory,
    #     name=f"temp_{split}"
    # )

    for img_file in os.listdir(img_dir):#for sample in samples:

        # sample_id = os.path.basename(sample.filepath).split(".")[0]

        sample_path = os.path.join(img_dir, img_file)
        sample_id = os.path.splitext(img_file)[0]

        # Create new sample
        sample = fo.Sample(filepath=sample_path)


        ann_file_label = os.path.join(lbl_dir, f"{sample_id}.txt")
        detections = []
        with open(ann_file_label, "r") as f:
          for line in f:
            parts = line.strip().split()
            xmin, ymin, w, h, cls = parts
            xmin, ymin, w, h = map(float, [xmin, ymin, w, h])
            cls = int(cls)

            cls_name=labelx[cls]

            det = fo.Detection(
                label=cls_name,
                bounding_box=[xmin, ymin, w, h]
                )

            detections.append(det)

        sample["ground_truth"] = fo.Detections(detections=detections)


        ann_file_pred = os.path.join(pred_dir, f"{sample_id}.txt")
        detections = []
        with open(ann_file_pred, "r") as f:
          for line in f:
            parts = line.strip().split()
            xmin, ymin, w, h, cf_sc, cls,anc_id = parts
            xmin, ymin, w, h, cf_sc = map(float, [xmin, ymin, w, h, cf_sc])
            cls = int(cls)

            cls_name=labelx[cls]

            anc_id = int(anc_id)

            det = fo.Detection(label=cls_name,
                               bounding_box=[xmin, ymin, w, h],
                               confidence=cf_sc,
                               anchor_id=anc_id
                               )

            detections.append(det)

        sample["prediction"] = fo.Detections(detections=detections)

        sample["split"] = split


        dataset.add_sample(sample)
        #sample.save()

# # Save
# dataset.persist()

```

```python
# Assume you already stored predictions in dataset with field "predictions"
# and ground truth is in "ground_truth"

results = dataset.evaluate_detections(
    "prediction",
    gt_field="ground_truth",
    eval_key="eval",
    iou=0.5,
    classwise=True,
    compute_mAP=True  #  important

)

#print("mAP:", results.mAP())
print("Per-class AP:", results.classes)

results.print_report()

print(f"mAP score: {results.mAP()}")
# for cls, ap in results.classwise_AP.items():
#     print(f"{cls}: {ap:.4f}")


# # Print a classification report for the top-10 classes
# results.print_report()



# --- 4. Plot interactive version with Plotly ---
# heat = results.plot_confusion_matrix(
#     classes=results.classes,
#     include_other=True,
#     include_missing=True,
#     backend="plotly"
#      # gives InteractiveHeatmap
# )
# heat.show()
#print(heat/)
#heat.figure.show()  # opens interactive Plotly plot

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = results.confusion_matrix()
print(cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=results.classes)
disp.plot()
plt.show()


#results.report()
#print(heat)
```
